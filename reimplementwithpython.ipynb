```python
import math
qs = [0.05, 0.10, 0.20, 0.30, 0.45]
print("Example 5.4 — Minimum z for ε = 10^-6 and ε = 10^-3")
print("Formula: z_min = ⌈ ln(ε) / ln(q/p) ⌉")
print(f"\n{'':>6}  {'':>12}  {'ε = 10^-6':>10}  {'ε = 10^-3':>10}")
print(f"{'q':>6}  {'ln(q/p)':>12}  {'z_min':>10}  {'z_min':>10}")
print("-" * 48)
for q in qs:
    p = 1 - q
    ln_ratio = math.log(q / p)
    z6 = math.ceil(math.log(1e-6) / ln_ratio)
    z3 = math.ceil(math.log(1e-3) / ln_ratio)
    print(f"{q:>6.2f}  {ln_ratio:>12.3f}  {z6:>10}  {z3:>10}")
print()
print("Verification: Q_z = (q/p)^z for each entry")
print(f"\n{'q':>6}  {'z':>4}  {'(q/p)^z':>14}  {'≤ ε?':>8}")
for q in qs:
    p = 1 - q
    for eps, label in [(1e-6, "1e-6"), (1e-3, "1e-3")]:
        z = math.ceil(math.log(eps) / math.log(q/p))
        Qz = (q/p)**z
        check = "✓" if Qz <= eps else "✗"
        print(f"{q:>6.2f}  {z:>4}  {Qz:>14.2e}  {check:>5} (ε={label})")
```

```python
def nakamoto_P(q, z):
    if z == 0:
        return 1.0
    p = 1.0 - q
    lam = z * q / p
    s = 1.0
    for k in range(z + 1):
        poisson = math.exp(-lam)
        for i in range(1, k + 1):
            poisson *= lam / i
        s -= poisson * (1 - (q / p) ** (z - k))
    return s
 

print("Example 5.6: q = 0.3, z = 5")
print("p = 0.7, λ = 15/7 ≈ 2.143")
 
q, p, z = 0.3, 0.7, 5
lam = z * q / p
 
print(f"\n{'k':<4} {'P_k':<10} {'(3/7)^(5-k)':<14} {'P_k(1-(3/7)^(5-k))':<22}")
print("-" * 52)
total = 0.0
for k in range(z + 1):
    Pk = lam**k * math.exp(-lam) / math.factorial(k)
    catch = (q / p) ** (z - k)
    contrib = Pk * (1 - catch)
    total += contrib
    print(f"{k:<4} {Pk:<10.5f} {catch:<14.5f} {contrib:<22.5f}")
 
P = 1 - total
print(f"\nSum = {total:.5f}, so P = 1 − {total:.5f} = {P:.5f}")
print(f"Matches Nakamoto [1]: P = 0.17735  ✓")
 
qs = [0.05, 0.10, 0.15, 0.20, 0.25, 0.30, 0.45]
zs = [0, 1, 3, 6, 10, 15, 24]
 
header = f"{'z':<5}" + "".join(f"{'q='+str(q):<10}" for q in qs)
print(f"\n{header}")
print("-" * 78)
 
for z in zs:
    row = f"{z:<5}"
    for q in qs:
        val = nakamoto_P(q, z)
        if val < 5e-4:
            row += f"{'–':<10}"
        else:
            row += f"{val:<10.3f}"
    print(row)


print("High-precision values (6 dp) for LaTeX cross-check")

 
header = f"{'z':<5}" + "".join(f"{'q='+str(q):<14}" for q in qs)
print(f"\n{header}")
print("-" * 90)
 
for z in zs:
    row = f"{z:<5}"
    for q in qs:
        val = nakamoto_P(q, z)
        row += f"{val:<14.6f}"
    print(row)
```
